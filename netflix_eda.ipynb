```python
rating_map = {
    'TV-PG': 8,
    'TV-MA': 9,
    'TV-Y7-FV': 7,
    'TV-Y7': 7,
    'TV-14': 6,
    'R': 9,
    'TV-Y': 8,
    'NR': 5,
    'PG-13': 8,
    'TV-G': 8,
    'PG': 8,
    'G': 6,
    None: None,
    'UR': 5,
    'NC-17': 9
}

# Map the ratings to their corresponding numeric values
df['numeric_rating'] = df['rating'].map(rating_map)
df
```

```python
def categorize_rating(numeric_rating):
    if numeric_rating >= 9:
        return 'Superhit'
    elif numeric_rating >= 7:
        return 'Hit'
    elif numeric_rating == 6:
        return 'Average'
    elif pd.isnull(numeric_rating): 
        return 'None'
    else:
        return 'Flop'

df['rating_category'] = df['numeric_rating'].apply(categorize_rating)
df
```

```python
ratings=df.groupby(["numeric_rating"]).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

```python
import re  
def convert_to_hours_minutes(duration_str):
    if pd.isna(duration_str):  
        return duration_str
    match = re.search(r'(\d+) min', duration_str)  
    if match:
        minutes = int(match.group(1))  
        hours = minutes // 60
        remaining_minutes = minutes % 60
        return f"{hours} hr {remaining_minutes} min"  
    else:
        return duration_str 

df['duration_converted'] = df['duration'].apply(convert_to_hours_minutes)
```
